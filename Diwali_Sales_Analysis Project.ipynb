# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
 sns.countplot(x = 'Gender',data = df) #countplot counts the value of x, data is dataframe which has the column x
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df) #sns is seaborn

for bars in ax.containers:
    ax.bar_label(bars) #also gives values to the bars
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False) #we want to do groupby on the basis of gender column and we want to sum amount columns by sorting on the basis of amount

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen) #sales_gen data is passed

# Bar plot with different colors for each bar
sns.barplot(
    x='Gender',
    y='Amount',
    data=sales_gen,
    palette='Set2'   # You can try 'Set1', 'pastel', 'coolwarm', etc.
)


plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender') #hue splits the column into two part with different colors

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
df.columns
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(16) #total number of orders state wise has to be observed

sns.set(rc={'figure.figsize':(27,6)}) #plot size has to be given so that values does not overlap
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)}) #if u don't write this, it will take the width of initial question
# Countplot with custom colors
ax = sns.countplot(
    data=df,
    x='Marital_Status',
    palette='Set2'   # Try 'Set1', 'pastel', 'husl', 'coolwarm', etc.
)
#adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender') #0 is married
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
# Get unique number of occupations
num_colors = df['Occupation'].nunique()
#Countplot with unique colors for each occupation
ax = sns.countplot(
    data=df,
    x='Occupation',
    palette=sns.color_palette('husl', num_colors)  # 'husl' gives distinct colors
)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()  # maximum number of orders are given by people working in IT sector
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')

# Get unique number of occupations
num_colors = sales_state['Occupation'].nunique()

# Barplot with unique colors for each occupation
sns.barplot(
    data=sales_state,
    x='Occupation',
    y='Amount',
    palette=sns.color_palette('husl', num_colors)  # distinct colors
)

plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(28,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
# Get unique number of occupations
num_colors = df['Product_Category'].nunique()
#Countplot with unique colors for each occupation
ax = sns.countplot(
    data=df,
    x='Product_Category',
    palette=sns.color_palette('husl', num_colors)  # 'husl' gives distinct colors
)

 
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
# Get unique number of product categories in this top 10
num_colors = sales_state['Product_Category'].nunique()

# Barplot with different colors for each product category
ax = sns.barplot(
    data=sales_state,
    x='Product_Category',
    y='Amount',
    palette=sns.color_palette('husl', num_colors)  # distinct colors
)



plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
# Get unique number of product ID in this top 10
num_colors = sales_state['Product_ID'].nunique()

# Barplot with different colors for each product category
ax = sns.barplot(
    data=sales_state,
    x='Product_ID',
    y='Orders',
    palette=sns.color_palette('husl', num_colors)  # distinct colors
)



plt.show()
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
